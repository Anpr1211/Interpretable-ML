```python
# importing libraries

import pandas as pd
import matplotlib.pyplot as plt
```

The 5 datasets are downloaded from www.dt.fee.unicamp.br/~tiago//youtubespamcollection/.

They are loaded and a look at the structure is given to ensure that it is the same for all the 5 datasets.

```python
psy = pd.read_csv("Youtube 01-comments Psy.csv")

psy.head()
```

```python
KatyPerry = pd.read_csv("Youtube 04-comments KatyPerry.csv")

KatyPerry.head()
```

```python
LMFAO = pd.read_csv("Youtube 07-comments LMFAO.csv")

LMFAO.head()
```

```python
Eminem = pd.read_csv("Youtube 08-comments Eminem.csv")

Eminem.head()
```

```python
Shakira = pd.read_csv("Youtube 09-comments Shakira.csv")

Shakira.head()
```

Since the structure is same throughout, the datasets are merged into one.

```python
data = [psy, KatyPerry, LMFAO, Eminem, Shakira]

df = pd.concat(data)
df.head()
```

```python
len(df)
```

The irrelevant columns are deleted.

```python
df = df.drop(["COMMENT_ID", "AUTHOR", "DATE"], axis=1)
```

```python
df.head()
```

```python
df["CLASS"].value_counts().plot(kind='bar')
```

There is no class imbalance, which is a suitable condition for modelling.

```python
# saving the dataset

df.to_csv('YouTube_Spam_Comments.csv')
```
